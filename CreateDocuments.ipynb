```python
import random

random.seed(2020)

from wordfreq import top_n_list, random_words
import numpy as np
```

```python
# Somehow similar languages.
LANGUAGES = {
    'es': 'Spanish',
    'en': 'English',
    'pt': 'Portuguese',
    'fr': 'French',
    'de': 'German',
    'it': 'Italian',
    'nl': 'Dutch'
}

LANG_CODES = list(LANGUAGES.keys())

# Population of native speakers in Millions.
POPULATION = {
    'es': 480,
    'en': 379,
    'pt': 221,
    'fr': 77.2,
    'de': 76.1,
    'it': 64.8,
    'nl': 23.1
}

#[POPULATION.set(lang, 100) for lang in POPULATION]
```

```python
# Words per-language.
words = {}
unique_words = set()

# Fetch a fix number of words per-language.
for lang in LANG_CODES:
    words[lang] = top_n_list(lang, 10000)

    # Collect unique words.
    unique_words = unique_words.union(words[lang])
    
# Turn uniue words into a list.
unique_words = list(unique_words)    
```

```python
# Numero de clases.
n = len(POPULATION)

# Numero de documentos.
M = 100000

# Parametros.
M_w = 0
M_k = np.zeros(n)

# Generate some random documents.
docs = []

# Crear documentos proporcionales a la poblacion.
for k in random.choices(list(range(n)), weights=list(POPULATION.values()), k=M):

    # Seleccionar 100 palabras al azar.
    doc = random.choices(list(enumerate(words[LANG_CODES[k]])), k=10)    
    docs.append((k, doc))

    # Add all the words.
    M_w += len(doc)

    # Mas documentos en una clase.
    M_k[k] += len(doc)
    
# Numero de palabras.
N = len(unique_words)
M_jk = np.zeros((N, n))
    
# Count 
for k, doc in docs:
    for j, _ in doc:        
        M_jk[j,k] += 1
 
# Fix for zero. Add all the words to all the languages.
M += n
M_w += N * n
M_k += N
M_jk += 1
```
